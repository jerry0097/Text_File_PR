```python
import pandas as pd
script = pd.read_csv(r"C:\Users\jason\Documents\Jerry\Studies\Jerry DSc\IS\Python for DS- Project\conv.txt", header = None,delimiter = ':')
script.columns = ['Name', 'Speech']
script.to_csv(r"C:\Users\jason\Documents\Jerry\Studies\Jerry DSc\IS\Python for DS- Project\conv2.txt", index = None)
script
```

```python
import numpy as np
name_1 = script["Name"].unique()
name_1
```

```python
df = script.sort_values("Name")
df['Speech'] = df.groupby(['Name'])['Speech'].transform(lambda x : ' '.join(x))
df = df.drop_duplicates()
```

```python
def to_split(txt):
    return txt.split()
df['Speech'] = df['Speech'].apply(lambda x : list(set(to_split(x))))
df = df.reset_index(drop=True)
df
```
